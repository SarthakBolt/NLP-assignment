```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df =pd.read_csv('new.csv',encoding= 'latin-1')
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
x= df['question']
y=df['label']
x_train,x_test, y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
v=CountVectorizer()
v.fit(x_train)
vec_x_train= v.transform(x_train).toarray()
vec_x_test= v.transform(x_test).toarray()
```

```python
v.vocabulary
```

```python
from sklearn.naive_bayes import MultinomialNB,GaussianNB, BernoulliNB
```

### **Multi-nomial Naive Bayes**

```python
m= MultinomialNB()
m.fit(vec_x_train,y_train)
print(m.score(vec_x_test,y_test))
```

**Accuracy of 89.21**%

```python
sample = input('ask a question:')
vec = v.transform([sample]).toarray()
m.predict(vec)
```
